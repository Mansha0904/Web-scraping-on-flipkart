```python
import requests
from bs4 import BeautifulSoup
import csv
Product_name=[]
Prices=[]
Description=[]
Reviews=[]
for i in range(0,453):
    url="https://www.flipkart.com/search?q=mobile+phones+under+30000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    custom_header={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36"}
    r=requests.get(url,headers=custom_header)
    
    soup = BeautifulSoup(r.text, 'lxml')
    box=soup.find("div",class_="QSCKDh dLgFEE")
    
    row=box.find_all('a',class_="k7wcnx")
    #print(row)
    for row in row:
        names=row.find("div",class_="RG5Slk")
        #print(names)
        for i in names:
            name=i.text.strip()
            Product_name.append(name)
        #print(Product_name)
        prices=row.find("div",class_="hZ3P6w DeU9vF")
        #print(prices)
        if prices is not None:
            for i in prices:
                price=i.text.strip()
        else:
            price="Not given"
        Prices.append(price)
        #print(Prices)
        descriptions=row.find_all("ul",class_="HwRTzP")
        for i in descriptions:
            description=i.text
            Description.append(description)
        reviews=row.find_all("div",class_="MKiFS6")
        if reviews is not None:
            for i in reviews:
                review=i.text
        else:
            review=0.0
        Reviews.append(review)
```

```python
import pandas as pd
df=pd.DataFrame({"Product_name":Product_name,"Prices":Prices,"Description":Description,"Reviews":Reviews})
print(df)
```
